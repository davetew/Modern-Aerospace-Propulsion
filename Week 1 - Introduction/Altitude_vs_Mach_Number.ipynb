# Altitude vs Mach Number for Selected Aircraft
Aircraft data were taken from a ChatGPT 5.0 query.

```python
import matplotlib.pyplot as plt

# Aircraft data
aircraft_data = [
    ("Wright Flyer (1903)", 120, 0.02),
    ("Spirit of St. Louis (1927)", 10000, 0.13),
    ("DC-3 (1935)", 8000, 0.25),
    ("B-17 Flying Fortress (1938)", 25000, 0.35),
    ("P-51 Mustang (1940)", 25000, 0.55),
    ("Boeing 707 (1957)", 35000, 0.82),
    ("Concorde (1969)", 60000, 2.00),
    ("SR-71 Blackbird (1966)", 85000, 3.20),
    #("Boeing 747 (1970)", 35000, 0.85),
    ("F-22 Raptor (1997)", 50000, 2.00),
    ("B-2 Spirit (1989)", 50000, 0.95),
    #("B-21 Raider (2023)", 50000, 0.95),
]

names = [a[0] for a in aircraft_data]
altitudes = [a[1] for a in aircraft_data]
mach_numbers = [a[2] for a in aircraft_data]

# Create scatter plot
plt.figure(figsize=(14, 8))
plt.scatter(mach_numbers, altitudes)

# Label each point
for i, name in enumerate(names):
    plt.text(mach_numbers[i] + 0.03, altitudes[i], name, fontsize=12, rotation=45)

plt.title("Cruise Altitude vs Cruise Mach Number of Major Aircraft (1903â€“2023)", fontsize=16, weight="bold")
plt.xlabel("Cruise Mach Number", fontsize=14)
plt.ylabel("Cruise Altitude (ft)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

plt.show()
```
